```python
apple_stock_market_price = 135
print(apple_stock_market_price)

apple_stock_bought = 20
apple_stock_purchase_price = 60
total_apple_stocks_value = apple_stock_bought * apple_stock_purchase_price
print(total_apple_stocks_value)

apple_stock_new_added = 10
apple_stock_bp_new = 145
total_apple_stocks_value_new = apple_stock_new_added * apple_stock_bp_new
print(total_apple_stocks_value_new)

apple_stock_sold = 8
apple_stock_sp = 120
total_apple_stocks_sold_value = apple_stock_sold * apple_stock_sp
print(total_apple_stocks_sold_value)

Total_apple_stocks_after_selling = apple_stock_bought + apple_stock_new_added - apple_stock_sold
print(Total_apple_stocks_after_selling)

Total_current_value_of_apple_stocks = Total_apple_stocks_after_selling * apple_stock_market_price
print(Total_current_value_of_apple_stocks)

average_price_per_stock = (total_apple_stocks_value + total_apple_stocks_value_new) / (apple_stock_bought + apple_stock_new_added)
print(average_price_per_stock)

total_gain_loss = Total_current_value_of_apple_stocks - total_apple_stocks_value - total_apple_stocks_value_new + total_apple_stocks_sold_value
print(total_gain_loss)

tax_rate_france = 0.3
tax_amount = total_gain_loss * tax_rate_france
print(tax_amount)

tax_exempt_belgium = 500
tax_rate_belgium = 0.40
tax_amount_belgium = max(0, (total_gain_loss - tax_exempt_belgium) * tax_rate_belgium)
print(tax_amount_belgium)
```

```python
stock_value_now = 135  # current stock value
stocks_bought_first = 20
price_first = 60  # price at which the first set of stocks were bought
stocks_bought_second = 10
price_second = 145  # price at which the second set of stocks were bought
stocks_sold = 8
price_sold = 120  # price at which stocks were sold
```

```python
total_stocks = stocks_bought_first + stocks_bought_second - stocks_sold
total_value = stock_value_now * total_stocks
print(f"Total value of the current Apple stocks in John’s portfolio: ${total_value:.2f}")
```

```python
total_spent_first = stocks_bought_first * price_first
total_spent_second = stocks_bought_second * price_second
average_price = (total_spent_first + total_spent_second) / (stocks_bought_first + stocks_bought_second)
print(f"Average purchase price of John’s Apple stocks: ${average_price:.2f}")
```

```python
remaining_first = stocks_bought_first - stocks_sold
pnl_remaining_first = (stock_value_now - price_first) * remaining_first
pnl_second = (stock_value_now - price_second) * stocks_bought_second
pnl_sold = (price_sold - price_first) * stocks_sold
total_capital_gain = pnl_remaining_first + pnl_second + pnl_sold
print(f"Total capital gain if John sells all his Apple stocks now: ${total_capital_gain:.2f}")
```

```python
tax_rate_france = 0.30
tax_france = total_capital_gain * tax_rate_france
tax_exempt_belgium = 500
tax_rate_belgium = 0.40
tax_belgium = max(total_capital_gain - tax_exempt_belgium, 0) * tax_rate_belgium
print(f"Taxes in France on his capital gains: ${tax_france:.2f}")
print(f"Taxes in Belgium on his capital gains: ${tax_belgium:.2f}")
```
